```python
import numpy as np
import matplotlib.pyplot as plt

# simple prime check
def is_prime(n):
    if n < 2:
        return False
    if n % 2 == 0 and n != 2:
        return False
    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
    return True

# get prime p >= 5
p_values = [p for p in range(5, 51) if is_prime(p)]  # adjust 51 to a larger range if needed
y_values = [p**2 + 2 for p in p_values]

# plot only prime points
plt.scatter(p_values, y_values, color="red", label=r"$p^2 + 2$ (prime $p \geq 5$)")
for p, y in zip(p_values, y_values):
    plt.text(p, y, str(p), ha='center', va='bottom')  # label primes

plt.xlabel("p")
plt.ylabel("y")
plt.title("Plot of $p^2 + 2$ for prime $p \geq 5$")
plt.legend()
plt.grid(True)
plt.show()
```
